```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title="Image", image=None, size=10):
    w,h=image.shape[0],image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

Arithmetic Operations

```python
image = cv2.imread('images.png',0)
imshow("Grayscale Image", image)
print(image)

# Create a matrix of ones, then multiply it by a scaler of 100 
# This gives a matrix with same dimesions of our image with all values being 100
M = np.ones(image.shape, dtype="uint8") * 100
```

```python
print(M)
```

Increasing Brightness

```python
added = cv2.add(image, M)
imshow("Increased Brightness", added)

add2 = image + M
imshow("Increased Brightness", add2)
```

Decreasing Brightness

```python
Sub = cv2.subtract(image, M)
imshow("Decreased Brightness", Sub)

sub = image - M
imshow("Decreased Brightness 2", sub)
```

Bitwise Operations and Masking

```python
square = np.zeros((300,300), np.uint8)
cv2.rectangle(square, (50,50), (250,250), 255, -2)
imshow("Square", square)

ellipse = np.zeros((300,300), np.uint8)
cv2.ellipse(ellipse, (150,150), (100,50), 30, 0, 180, 255, -1)
imshow("Ellipse", ellipse)
```

Experimenting with some bitwise operations such as AND, OR, XOR and NOT

```python
And = cv2.bitwise_and(square, ellipse)
imshow("AND", And)

bitwiseOR = cv2.bitwise_or(square, ellipse)
imshow("OR", bitwiseOR)

bitwiseXOR = cv2.bitwise_xor(square, ellipse)
imshow("XOR", bitwiseXOR)

bitwiseNot_square = cv2.bitwise_not(square)
imshow("NOT Square", bitwiseNot_square)
```
